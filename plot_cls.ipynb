```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nside = 64
lmax = 3*nside-1
matrix_cl_mean = np.load('matrix_cl_mean.npy')
matrix_cl_std = np.load('matrix_cl_std.npy')
KS_cl_mean = np.load('KS_cl_mean.npy')
KS_cl_std = np.load('KS_cl_std.npy')
ell_th, TT, EE, BB, TE = np.loadtxt('totcls_lensed_r0p1.txt', unpack=True)
ell = np.load('ell.npy')
```

```python
plt.figure(figsize=(15,8)) 
plt.plot(ell_th[:lmax], BB[:lmax], label='Theory BB')

plt.errorbar(ell, KS_cl_mean[3], yerr=KS_cl_std[3], fmt='o', label='KS BB')

plt.errorbar(ell, matrix_cl_mean[3], yerr=matrix_cl_std[3], fmt='o', label='matrix BB')

plt.yscale('log')
plt.grid()

plt.legend()
plt.xlabel('$\ell$')
plt.ylabel('$D_\ell$')
plt.title('BB of r=0.1 maps')
```
